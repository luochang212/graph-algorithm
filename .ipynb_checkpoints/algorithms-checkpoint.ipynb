## 5.链接预测 Link Prediction

链接：[link_prediction](https://networkx.org/documentation/latest/reference/algorithms/link_prediction.html)

### 资源分配指数 resource_allocation_index

$\sum_{w \in \Gamma(u) \cap \Gamma(v)} \frac{1}{|\Gamma(w)|}$

其中 $\Gamma(u)$ 表示点 u 的邻居的集合。资源分配指数其实就是两个点交集的点的度数的倒数之和。

```python
preds = nx.resource_allocation_index(G7)
for u, v, p in preds:
    print(f"({u}, {v}) -> {p:.8f}")
```

```python
# 可只算部分 node pairs
list(nx.resource_allocation_index(G7, [(2, 3), (2, 5)]))
```

```python
nx.draw(G7, with_labels=True)
```

### jaccard 系数 jaccard_coefficient

$\frac{|\Gamma(u) \cap \Gamma(v)|}{|\Gamma(u) \cup \Gamma(v)|}$

u 和 v 所有邻居去重集合的大小作分母，u 和 v 所有邻居的交集去重集合的大小作分子。也就是共有邻居占比越大，则 jaccard 系数越大。

```python
preds = nx.jaccard_coefficient(G7)
for u, v, p in preds:
    print(f"({u}, {v}) -> {p:.8f}")
```

### 共邻中心性 common_neighbor_centrality

> Return the CCPA score for each pair of nodes.
> 
> Compute the Common Neighbor and Centrality based Parameterized Algorithm(CCPA) score of all node pairs in ebunch.

u 和 v 的 CCPA 分定义如下：

$\alpha \cdot(|\Gamma(u) \cap \Gamma(v)|)+(1-\alpha) \cdot \frac{N}{d_{u v}}$

其中：
 
- $\alpha$ is parameter varies between [0,1]
- $N$ denotes total number of nodes in the Graph and 
- $d_{u v}$denotes shortest distance between u and v



```python
preds = nx.common_neighbor_centrality(G7)
for u, v, p in preds:
    print(f"({u}, {v}) -> {p}")
```

```python
nx.draw(G7, with_labels=True)
```

### cn_soundarajan_hopcroft

> Count the number of common neighbors of all node pairs in ebunch using community information.

计算时考虑了社区信息

$|\Gamma(u) \cap \Gamma(v)|+\sum_{w \in \Gamma(u) \cap \Gamma(v)} f(w)$

如果 u he v 在同一个社区 $f(w)$ 为 1 否则为 0

```python
G8 = nx.path_graph(6)
G8.nodes[0]["community"] = 0
G8.nodes[1]["community"] = 0
G8.nodes[2]["community"] = 0
G8.nodes[3]["community"] = 1
G8.nodes[4]["community"] = 1
G8.nodes[5]["community"] = 1
preds = nx.cn_soundarajan_hopcroft(G8)
dotted_edges = []
for u, v, p in preds:
    print(f"({u}, {v}) -> {p}")
    G8.add_edge(u, v)
    G8[u][v]['weight'] = p
    dotted_edges.append((u, v))
```

```python
pos = nx.shell_layout(G8)

# 画点
node_color = [G8.nodes[i]["community"] for i in G8.nodes]
nx.draw_networkx_nodes(G8, pos, node_color=node_color)
nx.draw_networkx_labels(G8, pos)

# 添加边权数值
edge_labels = nx.get_edge_attributes(G8, 'weight')
nx.draw_networkx_edge_labels(G8, pos, edge_labels=edge_labels)

# 分别画实线和虚线
nx.draw_networkx_edges(G8, pos, edgelist=[edge for edge in G8.edges() if edge not in dotted_edges], edge_color='k')
nx.draw_networkx_edges(G8, pos, edgelist=dotted_edges, edge_color='k', style='dotted')
plt.axis('off')
plt.show()
```

### 优先连接 preferential_attachment

$|\Gamma(u)||\Gamma(v)|$

> Compute the preferential attachment score of all node pairs in ebunch.

```python
preds = nx.preferential_attachment(G7)
for u, v, p in preds:
    print(f"({u}, {v}) -> {p}")
```

```python
nx.draw(G7, with_labels=True)
```

### adamic_adar_index

$\sum_{w \in \Gamma(u) \cap \Gamma(v)} \frac{1}{\log |\Gamma(w)|}$

看起来是资源分配指数 resource_allocation_index 的改进版

```python
preds = nx.adamic_adar_index(G7)
for u, v, p in preds:
    print(f"({u}, {v}) -> {p:.8f}")
```

## 6.Matching

链接：[matching](https://networkx.org/documentation/latest/reference/algorithms/matching.html)

### is_matching

Return True if `matching` is a valid matching of `G`

```python
G9 = nx.Graph([(1, 2), (1, 3), (2, 3), (2, 4), (3, 5), (4, 5)])
nx.is_maximal_matching(G9, {1: 3, 2: 4})  # using dict to represent matching
```

```python
nx.is_matching(G9, {(1, 3)})  # using set to represent matching
```

```python
nx.draw(G9, with_labels=True)
```

### maximal_matching

Find a maximal matching in the graph.

```python
sorted(nx.maximal_matching(G9))
```

### is_perfect_matching

Return True if matching is a perfect matching for G

A perfect matching in a graph is a matching in which exactly one edge is incident upon each vertex.

```python
G10 = nx.Graph([(1, 2), (1, 3), (2, 3), (2, 4), (3, 5), (4, 5), (4, 6)])
my_match = {1: 2, 3: 5, 4: 6}
nx.is_perfect_matching(G10, my_match)
```

```python
nx.draw(G10, with_labels=True)
```
