# 社区检测算法

链接：[community](https://networkx.org/documentation/stable/reference/algorithms/community.html)

```python
import networkx as nx
from networkx.algorithms import community
```

## 1. 常见社区检测算法

```python
G = nx.karate_club_graph()

# Choose a layout
layout = nx.spring_layout(G)

options = {
    "pos": layout,
    "cmap": 'viridis',
    "with_labels": True,
}
```

```python
# label_propagation: 基于 Label propagation
communities = community.label_propagation_communities(G)

# Set node colors
dct = {e: i for i, lst in enumerate(communities) for e in lst}
node_color = [dct[node] for node in G.nodes]

# Plot the graph
nx.draw(G, node_color=node_color, **options)
```

```python
# girvan_newman: 基于中心度
SLOTS = 0

communities = community.girvan_newman(G)
communities = list(communities)[SLOTS]

# Set node colors
dct = {e: i for i, lst in enumerate(communities) for e in lst}
node_color = [dct[node] for node in G.nodes]

# Plot the graph
nx.draw(G, node_color=node_color, **options)
```

```python
# greedy_modularity: 基于模块度
communities = community.greedy_modularity_communities(G)

# Set node colors
dct = {e: i for i, lst in enumerate(communities) for e in lst}
node_color = [dct[node] for node in G.nodes]

# Plot the graph
nx.draw(G, node_color=node_color, **options)
```

```python
# louvain: 基于 Louvain Community Detection Algorithm
communities = community.louvain_communities(G, seed=123)

# Set node colors
dct = {e: i for i, lst in enumerate(communities) for e in lst}
node_color = [dct[node] for node in G.nodes]

# Plot the graph
nx.draw(G, node_color=node_color, **options)
```

## 2. 是否接受边权作为输入

实验结果：带权图推荐使用 louvain

```python
G = nx.Graph()

G.add_edge(1, 2, weight=0.2)
G.add_edge(2, 3, weight=0.8)
G.add_edge(3, 4, weight=0.2)
G.add_edge(1, 4, weight=0.8)

G.add_edge(1, 5, weight=1)
G.add_edge(2, 5, weight=0.2)
G.add_edge(2, 6, weight=1)
G.add_edge(3, 6, weight=1)
G.add_edge(3, 7, weight=0.2)
G.add_edge(4, 7, weight=1)
G.add_edge(4, 8, weight=1)
G.add_edge(1, 8, weight=1)

pos = nx.shell_layout(G, nlist=[[1,2,3,4], [5,6,7,8]])
edge_labels = nx.get_edge_attributes(G, 'weight')

nx.draw(G, pos, with_labels=True)
_ = nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
```

```python
options = {
    "pos": pos,
    "cmap": 'viridis',
    "with_labels": True,
}
```

```python
# label_propagation: 基于 Label propagation
communities = community.label_propagation_communities(G)

# Set node colors
dct = {e: i for i, lst in enumerate(communities) for e in lst}
node_color = [dct[node] for node in G.nodes]

# Plot the graph
nx.draw(G, node_color=node_color, **options)
```
