## 1.Node2Vec

```python
import networkx as nx
from node2vec import Node2Vec

# Generate a random graph
G = nx.fast_gnp_random_graph(n=100, p=0.05)

# Initialize Node2Vec with the graph
node2vec = Node2Vec(G, dimensions=64, walk_length=30, num_walks=200, workers=4)

# Fit Node2Vec to the graph
model = node2vec.fit(window=10, min_count=1)

# Retrieve node embeddings
embeddings = {node: model.wv[node] for node in G.nodes()}

# Example of using the embeddings
print("Embedding for node 0:", embeddings[0])
```

```python
ggen = (e for e in G.nodes)
i, stop = 0, 2
for node in ggen:
    print(node, '->', [k for k, _ in G.adj[node].items()])
    if i >= stop:
        break
    i += 1

# G.adj[3]
```
